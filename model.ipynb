```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
df.head()
```

```python
y = df['diagnosis']
x = df.drop(['id', 'diagnosis', 'Unnamed: 32'], axis=1)
x.head()
```

```python
ax = sns.countplot(y, label="Count")
b, m = y.value_counts()
print("Number of Benign: ", b)
print("Number of Malign: ", m)
```

```python
x.describe()
```

```python
data_n = (x - x.mean()) / x.std()
for i in range(3):
    data_type = pd.concat([y, data_n.iloc[:, i*10:(i+1)*10]], axis=1)
    data_type.head()
    data = pd.melt(data_type, id_vars='diagnosis',
                   var_name='features',
                   value_name='value')
    plt.figure(figsize=(10,10))
    sns.violinplot(x='features', y='value', hue='diagnosis', data=data, split=True, inner='quart')
    plt.xticks(rotation=45)
```

```python
sns.jointplot(x['radius_mean'], x['perimeter_mean'], kind="regg", color="red")
```

```python
sns.set(style="white")
data = x[['radius_worst', 'perimeter_worst', 'area_worst']]
print(data.head())
g = sns.PairGrid(data, diag_sharey=False)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw=3)
```

```python
sns.set(style="whitegrid", palette="muted")
for i in range(3):
    data_type = pd.concat([y, data_n.iloc[:, i*10:(i+1)*10]], axis=1)
    data = pd.melt(data_type, id_vars='diagnosis',
                   var_name='features',
                   value_name='value')
    plt.figure(figsize=(10,10))
    sns.swarmplot(x='features', y='value', hue='diagnosis', data=data)
    plt.xticks(rotation=45)
```

```python
f, ax = plt.subplots(figsize=(18,18))
sns.heatmap(x.corr(), annot=True, linewidths=0.5, fmt='.1f', ax=ax)
```

```python
list1 = ['area_mean', 'perimeter_worst', 'concave points_mean', 'area_se', 'area_worst', 'concave points_worst']
x1 = x[list1]
x1.head()
```

```python
f, ax = plt.subplots(figsize=(14, 14))
sns.heatmap(x1.corr(), annot=True, linewidths=0.5, fmt='.1f', ax=ax)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x1, y, test_size=0.3, random_state=42)
x_train_N = (x_train - x_train.mean()) / (x_train.max() - x_train.min())
x_test_N = (x_test - x_test.mean()) / (x_test.max() - x_test.min())
rf_clf = RandomForestClassifier(random_state=43)
rf_clf = rf_clf.fit(x_train_N, y_train)
y_predicted = rf_clf.predict(x_test_N)

score = accuracy_score(y_test, y_predicted)
print("Accuracy is ", score)
cm = confusion_matrix(y_test, y_predicted)
sns.heatmap(cm, annot=True, fmt='d')
```
